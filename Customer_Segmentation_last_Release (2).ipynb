## Calculate RFM Scoring

The simplest way to create customer segments from an RFM model is by using **Quartiles**. We will assign a score from 1 to 4 to each category (Recency, Frequency, and Monetary) with 4 being the highest/best value. The final RFM score is calculated by combining all RFM values. For Customer Segmentation, you will use the df_rfm data set resulting from the RFM analysis.

<br>
**Note**: Data can be assigned into more groups for better granularity, but we will use 4 in this case.

### i. Creating the RFM Segmentation Table


1. Create two functions, one for Recency and one for Frequency and Monetary. For Recency, customers in the first quarter should be scored as 4, this represents the highest Recency value. Conversely, for Frequency and Monetary, customers in the last quarter should be scored as 4, representing the highest Frequency and Monetary values.

````python
import pandas as pd

# Assuming RFM is your DataFrame with columns 'Recency', 'Frequency', and 'Monetary'
df_rfm = pd.DataFrame(RFM)

# Define custom bin edges for 'Frequency'
bin_edges = [0, 1, 3, 5, 224]
df_rfm['F_Quartile'] = pd.cut(df_rfm['Frequency'], bins=bin_edges, labels=range(1, 5), include_lowest=True)

# Continue with the remaining quartiles
df_rfm['R_Quartile'] = pd.qcut(df_rfm['Recency'], q=4, labels=range(4, 0, -1))
df_rfm['M_Quartile'] = pd.qcut(df_rfm['Monetary'], q=4, labels=range(1, 5))

# Combine RFM values to calculate the RFM score
df_rfm['RFM_Score'] = df_rfm['R_Quartile'].astype(str) + df_rfm['F_Quartile'].astype(str) + df_rfm['M_Quartile'].astype(str)

# Display the resulting RFM scores
print(df_rfm[['Recency', 'Frequency', 'Monetary', 'F_Quartile']])
````

````python
import pandas as pd

class RFMCalculator:
    def __init__(self, rfm_data):
        self.rfm_data = pd.DataFrame(rfm_data)

    def calculate_recency_quartiles(self):
        self.rfm_data['R_Quartile'] = pd.qcut(self.rfm_data['Recency'], q=4, labels=range(4, 0, -1))
        return self.rfm_data[['CustomerID', 'R_Quartile']]

    def calculate_fm_quartiles(self):
        self.rfm_data['F_Quartile'] = pd.qcut(self.rfm_data['Frequency'], q=4, labels=False, duplicates='drop')
        self.rfm_data['F_Quartile'] = pd.cut(self.rfm_data['F_Quartile'], bins=4, labels=range(1, 5))

        self.rfm_data['M_Quartile'] = pd.qcut(self.rfm_data['Monetary'], q=4, labels=range(1, 5))
        return self.rfm_data[['CustomerID', 'F_Quartile', 'M_Quartile']]

    def merge_quartiles(self, recency_quartiles, fm_quartiles):
        merged_df = pd.merge(recency_quartiles, fm_quartiles, on='CustomerID')
        return merged_df

    def calculate_rfm_scores(self):
        self.rfm_data['RFM_Score'] = (
            self.rfm_data['R_Quartile'].astype(str) +
            self.rfm_data['F_Quartile'].astype(str) +
            self.rfm_data['M_Quartile'].astype(str))
        return self.rfm_data[['CustomerID', 'RFM_Score']]

    def calculate_all_together(self):
        recency_quartiles = self.calculate_recency_quartiles()

        # Calculate Frequency and Monetary quartiles separately
        fm_quartiles = self.calculate_fm_quartiles()

        rfm_scores = self.calculate_rfm_scores()

        # Merge recency and frequency quartiles
        merged_df = pd.merge(recency_quartiles, fm_quartiles, on='CustomerID')

        # Merge with RFM scores
        merged_df = pd.merge(merged_df, rfm_scores, on='CustomerID')

        return merged_df

    def rfm_level(self):
        self.rfm_data['RFM_Level'] = pd.cut(
            self.rfm_data['RFM_Score'].astype(int),
            bins=[0, 111, 222, 333, 444],
            labels=['Low', 'Mid', 'High', 'Very High']
        )
        return self.rfm_data[['CustomerID', 'RFM_Level']]

    def calculate_average_values_and_size(self):
        if 'RFM_Level' not in self.rfm_data.columns:
            self.rfm_level()
        grouped_data = self.rfm_data.groupby('RFM_Level').agg({
            'Recency': 'mean',
            'Frequency': 'mean',
            'Monetary': 'mean',
            'RFM_Level': 'size'
        }).rename(columns={
            'Recency': 'Average_Recency',
            'Frequency': 'Average_Frequency',
            'Monetary': 'Average_Monetary',
            'RFM_Level': 'Segment_Size'
        }).reset_index()
        return grouped_data

# Assuming you have a DataFrame named RFM
# Create an instance of the RFMCalculator
cls = RFMCalculator(RFM)
RFM_combined = cls.calculate_all_together()
print(RFM_combined)
````

## Clustering Implementation

Try to apply different clustering algoritms to gather clusters like you achieved in manual RFM analysis.

### i. Define the Optimal Number of Clusters

````python
RFM_no_cust = RFM.drop('CustomerID', axis=1)
````

````python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
from math import isnan

# Calculating Hopkins score to know whether the data is good for clustering or not.
def hopkins(X):
    d = X.shape[1]
    n = len(X)
    m = int(0.1 * n)
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
    rand_X = sample(range(0, n, 1), m)
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
    HS = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(HS):
        print(ujd, wjd)
        HS = 0
    return HS
````

````python
hopkins(RFM_no_cust)
````

````python
from sklearn.cluster import KMeans

n_clusters = []
for i in range(2,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(RFM_no_cust)
    n_clusters.append(kmeans.inertia_)

# plot elbow graph
plt.plot(range(2,11),n_clusters,marker='o');
````

````python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2, 11)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(RFM_no_cust)
    clusterlabels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(RFM_no_cust, clusterlabels)
    print(
        f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}"
    )
````
